## **开始绘图，了解Matplotlib的基础底层布局**

参考文献：[知乎文章](https://www.zhihu.com/question/268583720/answer/2654032224)

### **1.像的整体框架与包含关系**


```python
fig,ax = plt.subplots(
                    nrows=1, ncols=1,         # 画布分成2*2个子图
                    figsize=(8, 3),         # 画布大小
                    )
ax.plot(df.y, color="red", label="Curve1")    # 作图函数，这里以折线图为例
ax.set_title('单张图片（Figure中仅有一个Axis）的情况', fontsize=15)          # 子图标题
ax.set_xlabel('X1', fontsize=10)              # X轴名称
ax.set_ylabel('Y1', fontsize=10)              # Y轴名称
```

### **2.关于fig和axes的设置**

```python
fig,ax = plt.subplots(
                    nrows=2, ncols=2,         # 画布分成2*2个子图
                    figsize=(20, 12),         # 画布大小
                    sharex=True, sharey=True  # 所有子图轴共享
                    )

fig.subplots_adjust(
                    wspace=0.2,               # 同一行元素之间距离
                    hspace=0.2)               # 同一列元素之间距离

fig.suptitle('多张图片（Figure中有多个Axis）的情况', fontsize=20)    # 画布的大标题

ax[0][0].plot(df.y, color="red", label="Curve1")    # 作图函数，这里以折线图为例
ax[0][0].set_title('第一张图', fontsize=15)          # 子图标题
ax[0][0].set_xlabel('X1', fontsize=10)              # X轴名称
ax[0][0].set_ylabel('Y1', fontsize=10)              # Y轴名称
ax[0][0].legend()                                   # 图例

ax[0][1].plot(df.x, color="blue", label="Curve2")
ax[0][1].set_title('第二张图', fontsize=15)
ax[0][1].set_xlabel('X2', fontsize=10)
ax[0][1].set_ylabel('Y2', fontsize=10)
ax[0][1].legend()

ax[1][0].plot(df.y, color="orange", label="Curve3")
ax[1][0].set_title('第三张图', fontsize=15)
ax[1][0].set_xlabel('X3', fontsize=10)
ax[1][0].set_ylabel('Y3', fontsize=10)
ax[1][0].legend()

ax[1][1].plot(df.x, color="yellow", label="Curve4")
ax[1][1].set_title('第四张图', fontsize=15)
ax[1][1].set_xlabel('X4', fontsize=10)
ax[1][1].set_ylabel('Y4', fontsize=10)
ax[1][1].legend()
```
