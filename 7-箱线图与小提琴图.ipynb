# <center>绘制数值特征的箱线图与小提琴图</center>
<br><br><br>

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 创建一个示例数据集
df = pd.read_csv("seaborn-data\\tips.csv")
```

## 绘制数据集每一列特征

```python
fig, ax = plt.subplots(
                    nrows=1, ncols=1,         # 画布分成2*2个子图
                    figsize=(8, 3),         # 画布大小
                    )
sns.boxplot(data=df, ax=ax)  
```

```python
fig, ax = plt.subplots(
                    nrows=1, ncols=1,         # 画布分成2*2个子图
                    figsize=(8, 3),         # 画布大小
                    )
sns.violinplot(data=df, ax=ax)  # 绘制小提琴图
```

## 绘制指定列

```python
fig, ax = plt.subplots(
                    nrows=1, ncols=1,         # 画布分成2*2个子图
                    figsize=(8, 3),         # 画布大小
                    )
sns.boxplot(
        y="total_bill", 
        data=df,
        ax=ax
        )  
```

```python
fig, ax = plt.subplots(
                    nrows=1, ncols=1,         # 画布分成2*2个子图
                    figsize=(8, 3),         # 画布大小
                    )
sns.violinplot(
            y="total_bill", 
            data=df, 
            ax=ax
            )  # 绘制小提琴图
```

## 绘制某一列关于某类别

```python
fig, ax = plt.subplots(
                    nrows=1, ncols=1,         # 画布分成2*2个子图
                    figsize=(8, 3),         # 画布大小
                    )
sns.boxplot(
        x="day", y="total_bill", 
        data=df,
        ax=ax
        )  
```

```python

fig, ax = plt.subplots(
                    nrows=1, ncols=1,         # 画布分成2*2个子图
                    figsize=(8, 3),         # 画布大小
                    )
sns.violinplot(
            x="day", 
            y="total_bill", 
            data=df, 
            ax=ax)
```
