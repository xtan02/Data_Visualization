```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("seaborn-data\\iris.csv").drop("species", axis=1)
df.head()
correlation = df.corr("pearson")
```

```python
fig, ax = plt.subplots(
                    nrows=1, ncols=1,         # 画布分成2*2个子图
                    figsize=(8, 6),         # 画布大小
                    )
sns.heatmap(data=correlation, 
            annot=True, 
            cmap="OrRd", 
            square=True,
            xticklabels='auto',
            yticklabels='auto',
            ax=ax)
plt.xticks(rotation = 45)
```
