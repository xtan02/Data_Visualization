```python
# 导入Matplotlib库中的pyplot，在Python中需要使用的库都需要类似地导入
import matplotlib.pyplot as plt
import seaborn as sns

# 导入Pandas库，后面我们主要需要这个库来引入我们自己的数据
import pandas as pd

# Python默认不能添加中文字段，这里的代码可以实现添加中文字段
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams['axes.unicode_minus'] = False  # 正常显示负号
```

```python
# 读取自己的数据集：seaborn-data文件夹中的anscombe.csv文件
df = pd.read_csv("seaborn-data\\penguins.csv")

# 预览读入的文件
df.head(5)
```

```python
fig,ax = plt.subplots(
                    nrows=1, ncols=1,         # 画布分成2*2个子图
                    figsize=(6, 4),         # 画布大小
                    )

# 使用Seaborn的scatterplot函数绘制散点图，并通过ax参数指定使用Matplotlib的子图进行绘制
sns.scatterplot(data=df, 
                x="bill_length_mm",
                y="bill_depth_mm", 
                ax=ax
                )

# 可以继续在ax对象上使用Matplotlib的函数进行设置，如设置标题、坐标轴标签等
ax.set_title('Scatter Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# 显示图形
plt.show()
```

```python
fig,ax = plt.subplots(
                    nrows=1, ncols=1,         # 画布分成2*2个子图
                    figsize=(6, 4),         # 画布大小
                    )

# 使用Seaborn的scatterplot函数绘制散点图，并通过ax参数指定使用Matplotlib的子图进行绘制
sns.scatterplot(data=df, 
                x="bill_length_mm", 
                y="bill_depth_mm", 
                hue="sex", 
                palette='Set1', 
                ax=ax,
                )

# 可以继续在ax对象上使用Matplotlib的函数进行设置，如设置标题、坐标轴标签等
ax.set_title('Scatter Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# 显示图形
plt.show()
```

```python
fig,ax = plt.subplots(
                    nrows=1, ncols=2,         # 画布分成2*2个子图
                    figsize=(12, 5),         # 画布大小
                    )

# 使用Seaborn的scatterplot函数绘制散点图，并通过ax参数指定使用Matplotlib的子图进行绘制
sns.scatterplot(data=df, 
                x="bill_length_mm",
                y="bill_depth_mm", 
                ax=ax[0],
                )

sns.scatterplot(data=df, 
                x="bill_length_mm", 
                y="bill_depth_mm", 
                hue="sex", 
                palette='Set1', 
                ax=ax[1],
                )

fig.suptitle('散点图效果预览', fontsize=20) 
plt.savefig('result-of-Python\\1-散点图效果预览.png')
```
