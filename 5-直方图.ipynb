```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("seaborn-data\\tips.csv")
df.head()
```

## 普通直方图

```python
fig, ax = plt.subplots(
                    nrows=1, ncols=1,         # 画布分成2*2个子图
                    figsize=(5, 3),         # 画布大小
                    )
sns.histplot(
        data=df, 
        x="total_bill", 
        binwidth=1, 
        ax=ax
        )
```

## 添加核密度估计来平滑直方图，提供有关分布形状的补充信息

```python
fig, ax = plt.subplots(
                    nrows=1, ncols=1,         # 画布分成2*2个子图
                    figsize=(5, 3),         # 画布大小
                    )
sns.histplot(data=df, x="total_bill", binwidth=1, kde=True, ax=ax)
```

## 多变量直方图

```python
fig, ax = plt.subplots(
                    nrows=1, ncols=1,         # 画布分成2*2个子图
                    figsize=(5, 3),         # 画布大小
                    )
sns.histplot(data=df, 
             x="total_bill", 
             hue="sex", 
             binwidth=1, 
             ax=ax
             )
```
