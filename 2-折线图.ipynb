```python
# 导入Matplotlib库中的pyplot，在Python中需要使用的库都需要类似地导入
import matplotlib.pyplot as plt
import seaborn as sns

# 导入Pandas库，后面我们主要需要这个库来引入我们自己的数据
import pandas as pd

# Python默认不能添加中文字段，这里的代码可以实现添加中文字段
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams['font.size'] = 12  # 字体大小
plt.rcParams['axes.unicode_minus'] = False  # 正常显示负号
```

```python
# 读取自己的数据集：seaborn-data文件夹中的dowjones.csv文件
df = pd.read_csv("seaborn-data\\dowjones.csv")

# 使用Pandas将日期列转换为时间序列日期
df["Date"] = pd.to_datetime(df["Date"])

# 预览读入的文件
df.head(5)
```

```python
fig, ax = plt.subplots(
                    nrows=1, ncols=1,         # 画布分成2*2个子图
                    figsize=(8, 3),         # 画布大小
                    )
ax.plot(
    df["Date"], df["Price"],
    color="blue",
    label=None
    )
ax.set_title('时间序列折线图', fontsize=15)          # 子图标题
ax.set_xlabel('日期', fontsize=10)              # X轴名称
ax.set_ylabel('价格', fontsize=10)              # Y轴名称
plt.savefig('result-of-Python\\2-折线图效果预览.png')
```
