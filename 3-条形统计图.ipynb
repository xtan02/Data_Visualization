```python
# 导入Matplotlib库中的pyplot，在Python中需要使用的库都需要类似地导入
import matplotlib.pyplot as plt
import seaborn.objects as so
import seaborn as sns

# 导入Pandas库，后面我们主要需要这个库来引入我们自己的数据
import pandas as pd

# Python默认不能添加中文字段，这里的代码可以实现添加中文字段
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams['font.size'] = 12  # 字体大小
plt.rcParams['axes.unicode_minus'] = False  # 正常显示负号
```

```python
# 读取自己的数据集：seaborn-data文件夹中的dowjones.csv文件
df = pd.read_csv("seaborn-data\\penguins.csv")

# 预览读入的文件
df.head(5)
```

## 对于单个变量的纯粹条形统计图 

函数根据变量类别（数值or分类）来判断图的作法，你的x、y只是决定图是竖着画还是横着画

```python
fig, ax = plt.subplots(figsize=(10, 4))

sns.countplot(x='island', palette='Set2', data=df, ax=ax)
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(10, 4))

sns.countplot(y='island', palette='Set2', data=df, ax=ax)
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(10, 4))

sns.countplot(x='island', hue="sex", palette='Set2', data=df, ax=ax)
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(10, 4))

sns.countplot(y='island', hue="sex", palette='Set2', data=df, ax=ax)
plt.show()
```

```python
fig,ax = plt.subplots(
                    nrows=2, ncols=2,         # 画布分成2*2个子图
                    figsize=(15, 9),         # 画布大小
                    )

fig.subplots_adjust(wspace=0.4)        

sns.countplot(x='island', palette='Set2', data=df, ax=ax[0][0])
sns.countplot(y='island', palette='Set2', data=df, ax=ax[0][1])
sns.countplot(x='island', hue="sex", palette='Set2', data=df, ax=ax[1][0])
sns.countplot(y='island', hue="sex", palette='Set2', data=df, ax=ax[1][1])

fig.suptitle('条形统计图效果预览', fontsize=20) 
plt.savefig('result-of-Python\\3-条形统计图效果预览.png')
```
